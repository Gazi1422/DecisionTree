```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('loan.csv')
df
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna("Male",inplace=True)
df["Married"].fillna("Yes",inplace=True)
df["Dependents"].fillna(0,inplace=True)
df["Self_Employed"].fillna("No",inplace=True)
df["LoanAmount"].fillna(df["LoanAmount"].mean(),inplace=True)
df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mean(),inplace=True)
df["Credit_History"].fillna(0.0,inplace=True)
```

```python
df.isnull().sum()
df.info()
df.drop(columns=["Dependents"],inplace=True)
```

```python
X=df.iloc[:,1:11]#independent
y=df.iloc[:,11]#dependent
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X_labelencoder = LabelEncoder()
X.iloc[:,0] = X_labelencoder.fit_transform(X.iloc[:,0])
X.iloc[:,1] = X_labelencoder.fit_transform(X.iloc[:,1])
X.iloc[:,2] = X_labelencoder.fit_transform(X.iloc[:,2])
X.iloc[:,3] = X_labelencoder.fit_transform(X.iloc[:,3])
X.iloc[:,8] = X_labelencoder.fit_transform(X.iloc[:,8])
X.iloc[:,9] = X_labelencoder.fit_transform(X.iloc[:,9])
print (X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(criterion = "entropy",max_depth=2)
decision_tree.fit(X_train, y_train) # training to machine
print('Decision Tree Classifer Created')
```

```python
predictValues =decision_tree.predict(X_test) # test the machine 
predictValues
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
print(accuracy_score(y_test, predictValues))
```
